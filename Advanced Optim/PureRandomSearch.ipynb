```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def onemax(x):
    return(np.sum(x))
```

```python
def needle(x):
    if(np.sum(x) == np.shape(x)[0]):
        return(1)
    else:
        return(0)
```

```python
def pureRS(n):
    x = np.random.randint(0,2,n)
    cpt = 0
    fitness = [onemax(x)]
    while(onemax(x) != n):
        cpt+=1
        y=  np.random.randint(0,2,n)
        if(onemax(y) > onemax(x)):
            x=y
        fitness.append(onemax(x))
    return(cpt, fitness)
```

```python
cpt_1 , fitness_1 = pureRS(10)
cpt_2 , fitness_2 = pureRS(10)
```

```python
plt.plot(fitness_1)
plt.plot(fitness_2)
```

```python
cpt_list = []
dev_list = []
for n in range(2,15):
    cpt = np.zeros(11)
    for i in range(11):
        cpt[i]= pureRS(n)[0]
    cpt_list.append(np.mean(cpt))
    dev_list.append(np.std(cpt))
```

```python
plt.plot(cpt_list ,label="Mean")
plt.legend()
plt.figure()
plt.plot(dev_list , label="STD")
plt.legend()
```
