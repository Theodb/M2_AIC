```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from scipy import signal
from scipy import ndimage
from PIL import Image
```

Extraction of our image and conversion in gray-scale:

```python
image=Image.open('barbara.jpg').convert('LA')

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

img=mpimg.imread('barbara.jpg')

img_gray=rgb2gray(img)

plt.imshow(img_gray)
```

### Gradient filter

#### Horizontal changes:
    
To find our gradient in the X direction we will convolve our image in gray-scale with the following kernel:

kernel_x = array([-1,0,1],[-2,0,2],[-1,0,1])

#### Vertical changes:
    
To find our gradient in the Y direction we will convolve our image in gray-scale with the following kernel:

kernel_y = array([-1,-2,-1],[0,0,0],[1,2,1])

```python
kernel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
gradient_x = signal.convolve2d(img_gray, kernel_x)  
kernel_y = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
gradient_y = signal.convolve2d(img_gray, kernel_y)

plt.subplot(1,2,1)
plt.imshow(gradient_x)
plt.title('Gradient x filter')

plt.subplot(1,2,2)
plt.imshow(gradient_y)
plt.title('Gradient y filter')

plt.show()
```

### Sobel filter

To create our Sobel filter we need to compute the square root of our (gradient x)^2 + (gradient y)^2

It's equivalent to take the hypotenus of our gradient x and gradient y:

```python
sobel_filter=np.hypot(gradient_x,gradient_y)

sobel_image = sobel_filter.astype(np.uint8)

plt.imshow(sobel_image, cmap=plt.cm.gray)
plt.title('Sobel filter')
plt.show()
```

### Averaging filter

To create an average filter we can use a kernel of 3*3 and convolve it with our image.

The kernel (3*3) of the mean is the following one: array([[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]])

```python
kernel_mean = np.array([[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]])
averaging_filter = signal.convolve2d(img_gray, kernel_mean)

plt.imshow(averaging_filter)
plt.title('Averaging filter')
plt.show()
```

### Gaussian filter

To create a Gaussian filter, we retrieve the Gaussian kernel with a parametrable size (=kernlen) and then in the same way than before we convolve it with our image. 

```python
def kernel_gaussian(kernlen, std=3):
    """
    Returns a 2D Gaussian kernel array.
    """
    k_gaussian_1d = signal.gaussian(kernlen, std=std).reshape(kernlen, 1)
    k_gaussian_2d = np.outer(k_gaussian_1d, k_gaussian_1d)
    return k_gaussian_2d


kernlen=3 #Choose our kernel size (i.e if kernlen=3 we will have a 3*3 matrix) 

print ("Our Gaussian kernel ",kernlen,"*",kernlen,":\n")
print(kernel_gaussian(kernlen,))

gaussian_filter = signal.convolve2d(img_gray, kernel_gaussian(kernlen,))

plt.imshow(gaussian_filter)
plt.title('Gaussian filter')
plt.show()
```
